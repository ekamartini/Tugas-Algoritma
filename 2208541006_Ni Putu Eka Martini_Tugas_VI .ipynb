## 5.21.7 Pengelompokkan Data menggunakan Pandas

```python
#============================================================
#Mengelompokkan data menggunakan Pandas
#Perintah mean ()
#============================================================
bunga_iris.groupby(['sepal width (cm)']).mean()
#Kode diatas menggunakan metode 'groupby()' dari pandas untuk engelompokkan baris di Data Frame berdasarkan nilai dari kolom 'sepal width(cm)'
#kemudian metode 'mean()' untuk menghitung nilai rata-rata dari setiap kelompok
#Ketika kode diatas dijalankan, akan dihasilkan suatu data bru yang berisi nilai rata-rata dari setiap kolom 'bunga iris' yg sudah dikelompokkan berdasarkan nilai kolom 'sepal width (cm)'
```

```python
#============================================================
#Mengelompokkan data menggunakan Pandas
#Perintah mean () dengan mengelompokkan
#============================================================
bunga_iris.groupby(['sepal width (cm)']).mean().sort_values('sepal length (cm)', ascending=True)
#Menggunakan metode 'groupby()' untuk mengelompokkan baris dalam DataFrame berdasaekan nilai 'sepal width(cm)'
#Menggunakan metode 'mean()' untuk menghitung rata-rata dari setiap kelompok.
#Menggunakan metode 'sort_values()' untuk mengurutkan data hasil kelompokkan berdasarkan kolom 'sepal length (cm)' secara menaik
#ketika kode tersebut dijalankan, output yang dihasilkan adalah data baru yang berisi nilai rata-rata dar setiap kolom yang sudah dikelompokkan berdasarkan nilai pada kolom 'sepal width (cm)
#Kemudian, data diurutkan berdasarkan kolom 'sepal length(cm)' secara menaik
```

```python
#============================================================
#Mengelompokkan data menggunakan Pandas
#Perintah sum
#============================================================

bunga_iris.groupby(['sepal width (cm)']).sum()
#pada perintah diatas, dilakukan suatu pengelompokkan data pada variabel bunga_iris berdasarkan kolom 'sepal width (cm)'
#Selanjutnya dilakukan agregasi nilai dengan fungsi 'sum()' 
#dimana untuk setiap kelompok data yang dihasilkan dari pengelompokkan tersebut
#Output yang dihasilkan adalah suatu data hasil penjumlahan untuk setiap kolom dari data bunga_iris 
#dengan perhitungannya berdasarkan kelompok-kelompok yang dibentuk berdasarkan nilai unik pada kolom 'sepal width(cm)'
```

## 5.22.1 Grafik Sederhana dengan Matplotlib

Pada dasarnya Matplotlib berfungsi untuk memplot data menjadi grafik, histogram atau bentuk lainnya untuk mempermudah visualisasi data yang sedang diolah.

```python
#============================================================
#Latihan mengenai Matplotlib
#Properti pembuatan grafik
#============================================================
#Grafik yang dibuat akan menampilkan hubungan x dan y dalam koordinat kartesius

#diinisialisasikan variabel x dan y yang merupakan list yang berisi 9 nilai dari masing-masing variabel x dan y yang berkaitan dengan fungsi y = x^3
x = [-4,-3,-2,-1,0,1,2,3,4]
y = [-64,-27,-8,-1,0,1,8,27,64]
#Kemudian, diinisialisasikan juga variabel x1 dan y1 yang juga merupakan list yang berisi masing-masing anggota dari variabel x1 dan y1 
x1 = [1,2,3,4,5,6,7]
y1 = [-2,0,2,4,6,8,10]

#digunakan method "plt.figure" untuk menentukan ukuran dan resolusi gambar
plt.figure(figsize=(8,8), dpi =300)
#figsize digunakan untuk menentukan gambar dengan format inci. sehingga pada kode ini ukuran gambar yang akan dihasilkan adalah 8x8 inci
#dpi digunakan untuk menentukan resolusi gambar dalam piksel per inci. Sehingga, pada kode ini resolusi yang dihasilkan adalah 300 piksel per inci

#Kemudian ditentukan font, ukuran font, serta warna dari garis dan grafik
plt.plot(x,y,label='y=x^3',color='purple',linewidth=5,marker='.',markersize =10,markeredgecolor='orange',linestyle='-')
#label='y=x^3' : dugunakan untuk memberi label atau judul pada plot
#color : digunakan untuk memberi warna pada plot
#linewidth : digunakan untuk memberi ketebalan garis pada plot
#marker : digunakan untuk menandai titik pada plot dengan bentuk titik
#markersize : digunakan untuk menentukan ukuran dari marker
#markerredgecolor : digunakan untuk memberi warna pada tepi marker
#linestyle : digunakan untuk menentukan jenis garis pada plot
plt.plot(x1,y1)
#metode ini digunakan untuk membuat plot dari data pada variabel x1 dan y1
#plot ini tidak memiliki variabel lain seperti label ataupun color sehingga plot ini akan menggunakan nilai default untuk setiap parameter
plt.title ('Grafik y =x^3', fontdict={'fontname' : 'Times New Roman', 'fontsize' :18})
#Ditentukan font dan ukuran font pada judul grafik
#'Grafik y =x^3', adalah string yang akan ditampilkan sebagai judul grafik
#fontdict : digunakan untuk mengatur jenis font dan ukuran font dari judul grafik
#fontname : jenis fnt
#fontsize : ukuran font
plt.xlabel ('x', fontdict={'fontname':'Comic Sans MS','fontsize':16})
plt.ylabel ('y', fontdict={'fontname':'Comic Sans MS','fontsize':16})
#Kemudian, sumbu x dan y diberi label pada grafik
plt.xticks ([-4,-3,-2,-1,0,1,2,3,4])
plt.yticks ([-64,-27,-8,-1,0,1,8,27,64])
#Digunakan fungsi plt.ticks untuk menentukan skala sumbu x dan t yang akan ditampilkan pada grafik
#pada sumbu x ditampilkan angka dari -4 sampai 4
#pada sumbu y ditampilkan anhka dari -64 sampai 64
plt.legend()
#metode ini digunakan untuk menampilkan legenda pada grafik yang menunjukkan label garis untuk grafik pertaman
plt.savefig('grafik.png', dpi=300)
#metode ini digunakan untuk menyimpan grafik sebagai file gambar dalam format PNG dan nama Grafik "grafik.png" dengan resolusi 300dpi
plt.show()
#Metode ini digunakan untuk menampilkan grafik pada jendela hasil
```

## 5.22.4 Properti Pembuatan Histogram dengan Matplotlib

```python
#============================================================
#Latihan mengenai Matplotlib
#Properti pembuatan Histogram
#============================================================

#Diinisialisasikan variabel pekerjaan yang berisi list untuk menjadi label pada sumbu x
pekerjaan = ['PNS', 'Guru', 'Buruh', 'Programmer', 'Petani']
#diinisialisasikan juga variabel jumlah sebagai nilai pada sumbu y
jumlah = [180, 55, 25, 20, 80]

#digunakan method "plt.figure" untuk menentukan ukuran dan resolusi gambar
plt.figure(figsize=(8,4))
#figsize digunakan untuk menentukan gambar dengan format inci. sehingga pada kode ini ukuran gambar yang akan dihasilkan adalah 8x4 inci
#dpi digunakan untuk menentukan resolusi gambar dalam piksel per inci. Sehingga, pada kode ini resolusi yang dihasilkan adalah 300 piksel per inci

histogram = plt.bar(pekerjaan,jumlah,color = 'orange')
#Selanjutnya kita memanggil metode untuk membuat histogram
#Kita juga dapat menentukan warna pada histogram dengan menambahan parameter color
#Variabel histogram berisi sebuah list yang berisi bar chart yang telah dibuat.
pola = ['o','x','/','.','*']
#Kita juga dapat membuat sebuah list 'pola' yang berisi simbol-simbol yang digunakan sebagai pola pada bar chart

#kemudian, kita dapat mengatur pola pada setiap bar chart dengan menggunakan metode '.set.hatch()' pada masing-masing bar chart yang ada pada variabel'histogram'
histogram [0].set_hatch('o')
histogram [1].set_hatch('x')
histogram [2].set_hatch('/')
histogram [3].set_hatch('.')
histogram [4].set_hatch('*')
plt.show()
#kode diatas dipanggil untuk menampilkan histogram yang telah dibuat
```

```python
#============================================================
#Latihan mengenai Matplotlib
#Properti pembuatan Histogram (2)
#============================================================

#Diinisialisasikan variabel pekerjaan yang berisi list untuk menjadi label pada sumbu x
pekerjaan = ['PNS', 'Guru', 'Buruh', 'Programmer', 'Petani']
#diinisialisasikan juga variabel jumlah sebagai nilai pada sumbu y
jumlah = [180, 55, 25, 20, 80]

#digunakan method "plt.figure" untuk menentukan ukuran dan resolusi gambar
plt.figure(figsize=(8,4))
#figsize digunakan untuk menentukan gambar dengan format inci. sehingga pada kode ini ukuran gambar yang akan dihasilkan adalah 8x4 inci
#dpi digunakan untuk menentukan resolusi gambar dalam piksel per inci. Sehingga, pada kode ini resolusi yang dihasilkan adalah 300 piksel per inci

hist = plt.bar(pekerjaan,jumlah,color = 'silver')
#Selanjutnya kita memanggil metode untuk membuat histogram
#variabel hist sebagai objek dari diagram batang tersebut
#Kita juga dapat menentukan warna pada histogram dengan menambahan parameter color
#Variabel hist berisi sebuah list yang berisi bar chart yang telah dibuat.
pola = ['o','x','/','.','*']
#Kita juga dapat membuat sebuah list 'pola' yang berisi simbol-simbol yang digunakan sebagai pola pada bar chart


for bar in hist :
#digunakan loop 'for' untuk mengakses setiap bar pada histogram variabel hist
    bar.set_hatch(pola.pop(0))
#pada setiap bar, diatur pola dengan menggunakan set_hatch dan pola.pop(0)
#pola.pop(0) digunakan untuk mengambil dan menghapus elemen pertama dari list pola sehingga setiap bar akan memiliki pola yang berbeda
    
    
plt.show()
#kode diatas dipanggil untuk menampilkan histogram yang telah dibuat
```

```python
#============================================================
#Latihan mengenai Matplotlib
#Properti pembuatan Histogram (3)
#============================================================

#Diinisialisasikan variabel pekerjaan yang berisi list untuk menjadi label pada sumbu x
pekerjaan = ['PNS', 'Guru', 'Buruh', 'Programmer', 'Petani']
#diinisialisasikan juga variabel jumlah sebagai nilai pada sumbu y
jumlah = [180, 55, 25, 20, 80]

#digunakan method "plt.figure" untuk menentukan ukuran dan resolusi gambar
plt.figure(figsize=(8,4))
#figsize digunakan untuk menentukan gambar dengan format inci. sehingga pada kode ini ukuran gambar yang akan dihasilkan adalah 8x4 inci
#dpi digunakan untuk menentukan resolusi gambar dalam piksel per inci. Sehingga, pada kode ini resolusi yang dihasilkan adalah 300 piksel per inci
hist = plt.bar(pekerjaan,jumlah,color = 'pink')
#Selanjutnya kita memanggil metode untuk membuat histogram
#variabel hist sebagai objek dari diagram batang tersebut
#Kita juga dapat menentukan warna pada histogram dengan menambahan parameter color
#Variabel hist berisi sebuah list yang berisi bar chart yang telah dibuat.
pola = ['o','x','/','.','*']
#Kita juga dapat membuat sebuah list 'pola' yang berisi simbol-simbol yang digunakan sebagai pola pada bar chart

for x in [0,1,2,3,4]:
#merupakan alternatif dari kode sebelumnya yang menggunakan perulangan 'for' untuk memberikan pola berbeda pada setiap histogram
    hist[x].set_hatch(pola[x])
#dalam perulangan, setiap bar akan diakses dengan menggunakan indeks x pada list 'hist'
#setiap pola diambil dari list 'pola' menggunakan indeks yang sama
#selanjutnya, setiap bar diberikan pola yang sesuai menggunakan metode 'set_hatch()'
    
plt.show()
#kode diatas dipanggil untuk menampilkan histogram yang telah dibuat
```
